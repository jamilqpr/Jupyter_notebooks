#Calculate the correlation between 'median_income' and 'median_house_value'
**Purpose: Assess the relationship between income and house value**

```python
correlation = df['median_income'].corr(df['median_house_value'])
print(f'Correlation = {correlation}')
```

#**Correlation**

Correlation is a measure of the linear relationship between two variables. A correlation coefficient close to 1 indicates a strong positive relationship, while a value close to -1 suggests a strong negative relationship. A coefficient near 0 implies little to no linear correlation.

In our dataset, we calculated the correlation between 'median_income' and 'median_house_value' to understand if there is any relationship between the income levels of residents and the median house values in the area. The calculated correlation coefficient is approximately 0.673. This positive correlation suggests that as median income increases, median house values tend to increase as well.

This finding implies that areas with higher incomes generally have higher-priced houses, which aligns with our expectations. However, it's important to remember that correlation does not imply causation. Other factors may also influence house prices.

#Compute the standard deviation of 'total_rooms'
**Purpose: Assess the variability in the number of total rooms.**

```python
std_dev_rooms = df['total_rooms'].std()
print(f'SD = {std_dev_rooms}')
```

#**Standard Deviation**

Standard deviation is a measure of the dispersion or variability in a dataset. A higher standard deviation indicates greater variability, while a lower value suggests data points are closer to the mean.

In our analysis, we computed the standard deviation of the 'total_rooms' variable. The calculated standard deviation is approximately 2155.59. This value provides insights into the spread of the number of total rooms in the houses within our dataset. A higher standard deviation suggests that there is a wide range in the number of rooms, with some houses having significantly more or fewer rooms than the average.

Understanding standard deviation is crucial for assessing the variability in our data, which can help us make informed decisions and predictions based on the dataset's characteristics.

In summary, by examining correlation and standard deviation, we gain valuable insights into the relationships and variability present in our dataset, allowing us to make more informed data-driven decisions.

#Histogram of 'median_income'
**Purpose: Visualize the distribution of median income in the dataset.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='median_income', bins=30, kde=True)
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()
```

#Scatter Plot of 'median_income' vs. 'median_house_value'
**Purpose: Explore the relationship between median income and median house value.**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='median_income', y='median_house_value', alpha=0.5)
plt.title('Scatter Plot of Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```
